```python
x = np.linspace(-10, 10, 1000)
y2 = np.cos(x)

plt.scatter(x, y2, marker='o', label='cos(x)')
plt.legend()
plt.show()
```

```python
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
plt.plot(x, y1, color='tab:blue', label='sin(x)')
plt.legend()
plt.show()
```

```python
#Exercise 8 
```

```python
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(2)
axs[0].plot(x, y1, color='tab:blue', label='sin(x)')
axs[1].scatter(x, y2, marker='o', label='cos(x)')
plt.legend()
plt.show()
```
