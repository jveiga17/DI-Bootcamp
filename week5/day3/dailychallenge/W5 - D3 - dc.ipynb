```python
#Task 1
#1

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#2

x = np.linspace(0, 20, 100)
print(x)
```

```python
#3

y = x**2
print(y)
```

```python
#4, 5

plt.plot(x, y)
plt.title("Plot of X vs Y (y = x^2)")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.xlim(0, 20)  # Adjust x-axis limits
plt.ylim(0, 400)  # Adjust y-axis limits
```

```python
#1
file_path = "C:\\Users\\User\\Desktop\\datasets\\chipo.csv"
chipo = pd.read_csv(file_path)
```

```python
#2
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:-1]))
```

```python
#3, 4
plt.scatter(chipo['item_price'], chipo['quantity'])
plt.title('Relationship between Item Price and Quantity Ordered')
plt.xlabel('Item Price ($)')
plt.ylabel('Quantity Ordered')
```

```python
#Task 3 
```

```python
#1

item_quantities = chipo.groupby('item_name')['quantity'].sum().reset_index()
print(item_quantities)
```

```python
#2, 3

top5_items = item_quantities.sort_values(by='quantity', ascending=False).head(5)
#print(top5_items)

plt.bar(top5_items['item_name'], top5_items['quantity'])

plt.title('Top 5 Items by Total Quantity Ordered')
plt.xlabel('Item Name')
plt.ylabel('Total Quantity Ordered')
```
